 #2.4  Plot a pie chart for all the 'Trend'  You can use the groupby function with the trend column to group all days with the same trend into a single group before plotting the pie chart. From the grouped data, create a BAR plot of average & median values of the 'Total Traded Quantity' by Trend type. 

```python
new = data[['Trend','Date']].copy()
grouped_df = new.groupby('Trend')

    
def trending(n):
    if n=='Slight or No change':
        return 0
    elif n=='Slight positive':
        return 1
    elif n=='Slight negative':
        return -1
    elif n=="Positive":
        return 2
    elif n=="Negative":
        return -2
    elif n== 'Among top gainers':
        return 3
    elif n== 'Among top losers':
        return -3
    elif n=="Bull run":
        return 4
    elif n == 'Bear drop':
        return -4
data["Trending"]= data["Trend"].apply(trending) 
data
d= data["Trending"].value_counts(dropna=False)
print(d)    
```

```python
fig = plt.figure(figsize = (8,8))
values = [170, 92, 80, 75, 51, 12,12,2]
percent=[]
for i in values:
    perc_days= (i/494)*100 
    perc_days=round(perc_days,2)
    percent.append(perc_days)
print(percent)    
labels = ['Slight or No change','Slight positive','Slight negative','Positive','Negative','Among top gainers','Among top losers','Bull run','Bear drop']

plt.pie(values, labels= percent,counterclock=False, shadow=True)
plt.title('Trends')
plt.legend(labels,loc=2)
plt.show()
 



```

```python
import pandas as pd
data2=pd.read_csv('week2.csv')
data2.head()
```

```python
data2['moving_21']=data2['Close Price'].rolling(14).mean()
data2['std_dev']=data2['Close Price'].rolling(2).std()
data2['std_2']=data2['std_dev']*2
data3=data2.dropna()
data3['upper']=data3['moving_21']+data3['std_dev']
data3['lower']=data3['moving_21']-data3['std_dev']
data3.head()
```

```python
data3['upper'].plot.line(figsize=(15,5))
data3['lower'].plot.line(figsize=(15,5))
```
