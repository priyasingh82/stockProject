```python
a= data['Close Price'].pct_change()
data['Day_Perc_Change']= a*100
data['Day_Perc_Change'].iloc[0]=0
data
```

```python
def trend(n):
    if -0.5 < n < 0.5 :
        return('Slight or No change')
    elif 0.5 < n < 1 :
        return ('Slight positive')
    elif -0.5 < n < -1 :
        return('Slight negative')
    elif 1 < n< 3:
        return('Positive')
    elif -1 < n < -3:
        return('Negative')
    elif 3 < n < 7:
        return('Among top gainers')
    elif -3 < n < -7:
        return('Among top losers')
    elif n> 7 :
        return('Bull run')
    elif n < -7:
        return('Bear drop')
    
data["Trend"]=data['Day_Perc_Change'].apply(trend)  
data
        
```

```python
av=data.groupby(['Trend'])['Total Traded Quantity'].mean()
med=data.groupby(['Trend'])['Total Traded Quantity'].median()
print(av)
print(med)
```
